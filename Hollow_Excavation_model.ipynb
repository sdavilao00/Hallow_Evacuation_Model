## Comparison of Shallow slopes (slow) vs Steep slopes (fast)

### Shallow slopes

```python
# lets find qs given that slope varies with hallow slope
```

```python
# Variables for Excavation model 
#infill_rate = 2  # in meters per yr
hollow_length = 10 #in meters
#hollow_width = 8 # in meters

#total_volume = infill rate * hollow_depth *hollow_width
```

```python
#find critical thickness
C = 2000 #Pa
pw = 1000 # kg/m3
ps = 2000 # kg/m3
phi = np.deg2rad(33)  #converts phi to radians 
g = 9.8  # m/s2
angle = int(input("angle in degrees "))
theta = np.deg2rad(angle) # converts theta to radians
```

```python
z_crit_shallow = C / ((pw*g**(np.tan(phi))*(np.cos(theta)))+(ps*g*(np.cos(theta))*(np.tan(theta) - np.tan(phi))))
print(z_crit_shallow)
```

```python
# now to find how long it takes to get to this critical depth
```

```python
Qin = 0.002  #m2/yr
Qout = 0.0005  #m2/yr
P = 0.0001  #m/yr
width = int(input("hollow width "))
time_shallow = z_crit_shallow / (Qin - Qout + (P*width))
print(time_shallow)
```

```python
#calculate the volume
volume_shallow = z_crit_shallow * width * hollow_length
print(volume_shallow)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

## Calculate critical soil depth for a range of slopes
C = 2000 #Pa
pw = 1000 # kg/m3
ps = 2000 # kg/m3
g = 9.8  # m/s2
phi = np.deg2rad(33)

#establish an array of slope angles
slope = np.array(np.deg2rad(range(1,61)))

# create an array for critical soil depth at angles 1-60 
z_crit = []
for k in slope:
    z_value = C / ((pw*g**(np.tan(phi))*(np.cos(k)))
                          +(ps*g*(np.cos(k))*(np.tan(k) 
                                                  - np.tan(phi))))
    z_crit += [z_value]

## Create a dataframe to store these values
hollow_data = pd.DataFrame(zip(slope, z_crit))

# rename column names
hollow_data.columns = ['gradient', 'critical_soil_depth']

# now make a quick plot about this
#plt.figure()  #this is an empty canvas
plt.plot(hollow_data.gradient, hollow_data.critical_soil_depth)
plt.xlabel('Gradient')
plt.ylabel('Depth (m)')
plt.grid()
plt.show()
```
